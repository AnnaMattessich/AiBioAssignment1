Write a Python function, called `find_max()`, to find and return the largest number in a non-empty list of numbers. E.g., if `nums = [1, 2, 3, 4, 5]`, `find_max(nums)` should return `5`. 

_Note:_ Python has a builtin `max()` function; **don't use it for this exercise!** The goal here is to practice solving basic programming problems on your own.

_Purpose:_ To give you practice solving basic programming problems with Python and working with array-like data structures.

```python
##import libraries
import pandas as pd
import numpy as np

##define the function find_max
def find_max(arr):
    ##sort the list so it is largest to smallest
    sorted_arr = np.sort(arr, kind='quicksort')[::-1]
    ##take the first value from the sorted array
    max_value = sorted_arr[0]
    ##return this value
    return max_value
```

Create and test some sample lists to demonstrate that your `find_max()` function works correctly.

```python
##example list 
list1 = [1, 3, 5, 7, 9, 1]

print("Maximum value in the array:", find_max(list1))
```

## Problem 3--All students (10 points)

**3a**: Using the `world_population_data.csv` file, make a plot showing the population of a country of your choosing over time from 1970 to 2023. 


```python
# Your code here for 3a:
import matplotlib.pyplot as plt
import pandas as pd

csv_file_path = 'world_population_data.csv'
data_ = pd.read_csv(csv_file_path)
data2 = pd.DataFrame(data_)
selected_row = data2.iloc[0]

years = selected_row.index[4:13] 
population = selected_row.values[4:13]

plt.figure(figsize=(15, 8))
plt.plot(years, population, marker='o', linestyle='-', color='b')
plt.title(f"Population Change from 1970 to 2023 - {selected_row['country']}")
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

**3b**: Generate some other plot with these data. This can be whatever you want to plot and can be whatever format. Have fun exploring the data and see what you can come up with!

```python
# Your code here for 3b:
continent = data2['continent']
population_23 = data2['2023 population']

continent_pop = pd.DataFrame({'continent': continent,
                              'population_23': population_23})

sorted_data = continent_pop.sort_values(by='continent')

sum_by_continent = sorted_data.groupby('continent')['population_23'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(sum_by_continent['continent'], sum_by_continent['population_23'], color='skyblue')
plt.title('Total Population by Continent in 2023')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.show()
```
